## Коллокации и Pointwise Mutual Information (PMI)

**Коллокация** - сочетание из двух или более слов, которое:

* обладает некомпозициональной семантикой
* состоит из двух слов, которые встречаются вместе значимо часто

Выявление колокаций (статистически значимых комбинаций слов) помогает лучше интерпретировать текст, а также находить именованные сущности

Поэтому, важно уметь выделять их автоматически 

Сначала мы научимся выделять коллокации, состоящие из двух слов --  **биграммы**

**[Pointwise Mutual Information (PMI)](https://en.wikipedia.org/wiki/Pointwise_mutual_information)**

$$ PMI(x, y) = log \frac{{p(x,y)}}{p(x) \cdot p(y)} $$

* x и y - слова, входящие в биграмму
* p(x,y) - вероятность появления биграммы x + y
* p(x) и p(y) -- вероятность появления каждого из элементов биграммы в отдельности

Для подсчёта PMI нужно сделать следующие шаги:

* предобработать текс (отчистить от мусора, нормализовать)
* составить частотный словарь слов в тексте
* найти вероятность появления каждого слова в тексте, разделив его частотность на общее количество слов
* сделать частотный список всех биграмм в тексте
* найти вероятность появления каждой биграммы в тексте, разделив её частотность на общее количество биграмм (а сколько их?)
* найти pmi по формуле
* биграммы с наибольшим pmi - коллокации в этом тексте
